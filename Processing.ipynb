```python
import pandas as pd
from scipy.constants import G
from astropy.constants import M_earth, R_earth 
from numpy import sqrt
```

```python
df = pd.read_excel("SpaceObjects.xlsx")
df.dropna(subset=df.columns, inplace=True)
df.head()
```

```python
df.info()
```

```python
df["Altitude"] = (df["Apogee"] + df["Perigee"]) / 2
df["Velocity"] = sqrt(G * M_earth / (R_earth / 1000 + df["Altitude"]))
df["Velocity"] = df["Velocity"].round(2)
df.head()
```

```python
df.to_excel("data.xlsx")
```
